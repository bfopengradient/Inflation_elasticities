### Australian inflation elasticities vs unemployment and short term interest rates.

The variables of interest were as follows:

1. Inflation
2. Unemployment rate
3. Short term interest rates 

The data ranges from September 1998 to March 2018. It is acknowledged that this analysis is simplistic but the idea is to get a visualisation of the rolling elasticities over time. Both charts plot the R2, the elasticity and the p-value of the elasticity over time.

#### Chart 1. 
The first chart shows the changing relationship between Australian Inflation and Unemployment since 2008. What used to be a statistically significant negative elasticity, where a drop in unemployment would imply an increase in inflation, has given way to a slightly positive or flat elasticity post 2008. As unemployment rates dropped post 2008 so too did inflation.  

#### Chart 2. 
The second chart helps illustrate the Central Bank's struggle to generate inflation within its target band. The elasticity of inflation against lagged short term interest rates has remained low since 2008. There are many quarters onwards from 2008 where the central bank has kept rates low or actually reduced rates and the elasticity between lagged short term rates and inflation has remained stubbornly low.
